```python
import pandas as pd
```

```python
f = open("chise.txt", "r")
unicode = []; char = []; component = []
for line in f.readlines():

    unicode.append(line[0:6])
    char.append(line[7])
    component.append(line[9:])
```

```python
data = pd.DataFrame({'unicode': unicode,
             'character':char,
              'component': component
             
             })
data
```

```python
def removing_line_ending(df):
    result = df.replace('\n','')
    result = result.replace('\t', '')
    return result
data.component = data.component.apply(removing_line_ending)
```

```python
data.head(30)
```

```python
a = '⿱⿰&CDP-895C;&CDP-895C;一'
b = '龹@apparent=⿱丷夫'
def split(element):
    
    element = element.replace('@apparent=', '')
    element = element.replace('⿰', '')
    element = element.replace('⿱', '')
    element = element.replace('⿸', '')
    element = element.replace('⿺', '')
    element = element.replace('⿵', '')
    element = element.replace('⿴', '')
    element = element.replace('⿳', '')
    element = element.replace('⿹', '')
    element = element.replace('⿲', '')
    element = element.replace('⿷', '')
    element = element.replace('⿻', '')
    element = element.replace('⿶', '')
    
    result = []
    tmp = ''
    search_status = False
    for char in element:
        
        if char == '&':
            search_status = True
            tmp = ''
            continue
        elif char == ';':
            search_status = False
            result.append(tmp)
            continue
            
        if search_status:
            tmp += char
        else:
            result.append(char)
    return result

split(b)
```

```python
data['radical_1'] = data.component.apply(split)
data
```

```python
data['radical_1_len'] = data.radical_1.apply(len)
```

```python
data[data.character == '气']['radical_1'].values[0]
```

```python
def full_radicals(ori_radicals):
    global data
    result = []
    
    characters = data.character.values
    for radical in ori_radicals:
        if radical not in characters:
            result.append(radical)
        else:
            full_radical = data[data.character == radical]['radical_1'].values[0]
            result += full_radical
            
    return result
data['full_radicals'] = data.radical_1.apply(full_radicals)
```

```python
data['final_len'] = data['full_radicals'].apply(len)
```
